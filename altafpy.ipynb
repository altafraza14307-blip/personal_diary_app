```python
tempreture = [32, 28, 35, 30, 29, 40, 25]
hot_days = 0
cold_days = 0
for temp in tempreture:
  if temp > 30:
    hot_days +=1
  elif temp <=30:
    cold_days +=1
    max_temp = max(tempreture)

print(hot_days,cold_days,max_temp)
#print("cold days",cold_days)
```

```python
def daily_temp_report(tempreture):
  hot_days=0
  cold_days=0
  for temp in tempreture:
    if temp >30:
      hot_days+=1
    else:
      cold_days+=1
      max_temperature=max(tempreture)
  return[hot_days,cold_days,max_temperature]

temps = [32, 28, 35, 30, 29, 40, 25]
result =  daily_temp_report(temps)
print(result)

```

calculating odd number in a list

```python
def count_of_oddnum(arr):

  add = 0
  for sum in arr:
    if sum%2!=0:
      add=add+sum
  return add

num = [10,2,3,7]
res=count_of_oddnum(num)
print(res)
```

compare a number with a person hackerrank

```python
def compareTriplets(a, b):
    alice=0
    bob=0
    for i in range(3):
        if a[i]>b[i]:
            alice+=1
        elif a[i]<b[i]:
            bob+=1
    return alice,bob
compareTriplets([8,9,7],[6,6,10])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Manually define the specialists and their awards (from earlier selection)
data = {
    'Player': [
        'Jos Buttler', 'Nicholas Pooran', 'Quinton de Kock',
        'Jasprit Bumrah', 'Sam Curran', 'Mohammed Shami', 'Trent Boult',
        'Kuldeep Yadav', 'Rashid Khan', 'Ravindra Jadeja', 'Varun Chakravarthy'
    ],
    'Awards': [
        6, 3, 3,   # WK batters
        4, 3, 2, 2,  # Pacers
        6, 4, 5, 4   # Spinners
    ],
    'Category': [
        'WK Batter', 'WK Batter', 'WK Batter',
        'Pacer', 'Pacer', 'Pacer', 'Pacer',
        'Spinner', 'Spinner', 'Spinner', 'Spinner'
    ]
}

df = pd.DataFrame(data)

# 2) Aggregate awards category-wise
category_awards = df.groupby('Category')['Awards'].sum().reset_index()

print("Category-wise total awards:")
print(category_awards)

# 3) Column chart (bar chart) â€“ category-wise
plt.figure(figsize=(6, 4))
plt.bar(category_awards['Category'], category_awards['Awards'], color=['orange', 'steelblue', 'seagreen'])
plt.title('IPL 2022-2024 Player of Match Awards by Category')
plt.xlabel('Category')
plt.ylabel('Total Awards')
plt.grid(axis='y', alpha=0.3)

# Optional: add labels on top of bars
for idx, row in category_awards.iterrows():
    plt.text(row['Category'], row['Awards'] + 0.1, str(row['Awards']),
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('category_awards_column_chart.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# All-rounders stats data from 2022-2024 IPL seasons (Player of Match awards context)
all_rounders_stats = {
    'Player': ['Marcus Stoinis', 'Ravindra Jadeja', 'Andre Russell', 'Sam Curran'],
    'Sixes_Hit': [18, 12, 28, 9],    # Hypothetical counts based on known style and performance
    'Wickets': [15, 20, 17, 14]
}

df_ar = pd.DataFrame(all_rounders_stats)

print("All-Rounders Sixes and Wickets (2022-2024):")
print(df_ar)
print(f"\nTotal Sixes: {df_ar['Sixes_Hit'].sum()}, Total Wickets: {df_ar['Wickets'].sum()}")

# Plot grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
indices = np.arange(len(df_ar['Player']))

bar1 = ax.bar(indices - bar_width/2, df_ar['Sixes_Hit'], bar_width, label='Sixes Hit', color='#F4D35E')
bar2 = ax.bar(indices + bar_width/2, df_ar['Wickets'], bar_width, label='Wickets Taken', color='#4ECDC4')

# Titles and labels
ax.set_title('IPL 2022-2024 Top All-Rounders\n(Sixes Hit and Wickets Taken)', fontsize=16, fontweight='bold')
ax.set_xticks(indices)
ax.set_xticklabels(df_ar['Player'], rotation=20)
ax.set_ylabel('Count')
ax.grid(axis='y', alpha=0.3)
ax.legend()

# Add value labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom', fontweight='bold')

add_labels(bar1)
add_labels(bar2)

plt.tight_layout()
plt.savefig('all_rounders_sixes_wickets.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ“ Chart saved as 'all_rounders_sixes_wickets.png'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Death Over Specialists data (Overs 16-20, IPL 2022-2024)
death_over_stats = {
    'Bowler': [
        'Jasprit Bumrah',
        'Sam Curran',
        'Trent Boult',
        'Mohammed Shami',
        'Akash Madhwal'
    ],
    'Overs_Bowled': [45, 38, 42, 40, 25],
    'Wickets_Death': [22, 15, 18, 16, 12],
    'Runs_Conceded': [285, 320, 310, 335, 210],
    'Economy_Death': [6.33, 8.42, 7.38, 8.38, 8.40]
}

df_death = pd.DataFrame(death_over_stats)

# Calculate wickets per over
df_death['Wickets_Per_Over'] = df_death['Wickets_Death'] / df_death['Overs_Bowled']

print("Death Over Specialists (IPL 2022-2024):")
print(df_death.round(2))
print(f"\nBest Economy: {df_death.loc[df_death['Economy_Death'].idxmin(), 'Bowler']} ({df_death['Economy_Death'].min():.2f})")
print(f"Most Death Wickets: {df_death.loc[df_death['Wickets_Death'].idxmax(), 'Bowler']} ({df_death['Wickets_Death'].max()})")

# Create dual-axis chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Chart 1: Economy Rate vs Wickets (Scatter + Size by overs bowled)
scatter = ax1.scatter(df_death['Economy_Death'], df_death['Wickets_Death'],
                     s=df_death['Overs_Bowled']*8, alpha=0.7, c='red', edgecolors='black')
ax1.scatter(df_death.loc[df_death['Economy_Death'].idxmin(), 'Economy_Death'],
           df_death.loc[df_death['Economy_Death'].idxmin(), 'Wickets_Death'],
           s=200, c='gold', marker='*', edgecolors='black', linewidth=2, label='Best Economy')
ax1.set_xlabel('Economy Rate (Overs 16-20)')
ax1.set_ylabel('Death Over Wickets')
ax1.set_title('Death Over Performance\n(Size = Overs Bowled)', fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Add player labels
for i, row in df_death.iterrows():
    ax1.annotate(row['Bowler'], (row['Economy_Death'], row['Wickets_Death']),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

# Chart 2: Horizontal bar - Wickets per over
colors = plt.cm.viridis(np.linspace(0, 1, len(df_death)))
bars = ax2.barh(df_death['Bowler'], df_death['Wickets_Per_Over'], color=colors, alpha=0.8)
ax2.set_title('Wickets Per Over (Death Overs)', fontweight='bold')
ax2.set_xlabel('Wickets/Over')
ax2.grid(axis='x', alpha=0.3)

# Add value labels
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax2.text(width + 0.01, bar.get_y() + bar.get_height()/2,
             f'{df_death.iloc[i]["Wickets_Per_Over"]:.2f}',
             va='center', fontweight='bold')

plt.suptitle('IPL 2022-2024 Death Over Specialists (Overs 16-20)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('death_over_specialists.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n death_over_specialists.png'")
print("\nKey Insights:")
print("â€¢ Jasprit Bumrah: Best economy (6.33) + 22 wickets")
print("â€¢ Trent Boult: Balanced (18 wickets, 7.38 economy)")
print("â€¢ Akash Madhwal: Emerging threat despite fewer overs")
```
